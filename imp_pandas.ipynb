```python
np.random.seed(20)
df = pd.DataFrame(np.random.randn(10,4),index=pd.date_range('2022-11-27',periods=10),columns=['a','b','c','d'])
```

```python
def test(x):
  if(x<0):
    return x
  else:
    return np.log(x)
```

```python
df["new_column_square_log"]=df['d'].apply(test)
```
