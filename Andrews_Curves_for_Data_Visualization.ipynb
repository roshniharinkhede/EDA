```python
url = 'https://raw.githubusercontent.com/nishant-harinkhede12/EDA-Projects/main/iris.csv'
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(url)
```

```python
df.info()
     
```

```python
plt.figure(figsize=(12,6))
pd.plotting.andrews_curves(df, 'species', colormap='YlOrRd')
```
